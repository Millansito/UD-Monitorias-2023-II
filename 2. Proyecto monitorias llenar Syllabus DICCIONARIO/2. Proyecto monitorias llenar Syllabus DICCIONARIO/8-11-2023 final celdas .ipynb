```python
import json
import openpyxl
from google.colab import files

# Subir archivo Excel
uploaded_excel = files.upload()
excel_file_name = list(uploaded_excel.keys())[0]  # Obtiene el nombre del archivo Excel

# Subir archivo JSON
uploaded_json = files.upload()
json_file_name = list(uploaded_json.keys())[0]  # Obtiene el nombre del archivo JSON

# Cargar el archivo JSON
with open(json_file_name, "r") as file:
    json_data = json.load(file)

# Cargar el archivo Excel con openpyxl
workbook = openpyxl.load_workbook(excel_file_name)
sheet = workbook.active

# Variables para rastrear si se encontró la palabra clave
found_keyword = False

# Recorre todas las celdas y realiza los reemplazos
for row in sheet.iter_rows(values_only=True):
    for cell in row:
        if not found_keyword:
            # Verificar si se encuentra la palabra clave en la celda
            if cell.value in json_data:
                keyword = cell.value
                found_keyword = True

        if found_keyword:
            if cell.value is None:
                key = keyword  # Usar la palabra clave como clave para buscar en el JSON
                if key in json_data:
                    cell.value = json_data[key]

# Guardar el archivo Excel modificado
workbook.save("excel_modificado.xlsx")

# Descargar el archivo Excel modificado
files.download("excel_modificado.xlsx")
```
