```python
# Importamos las bibliotecas necesarias para Google Colab
from google.colab import files

# Instalamos la biblioteca openpyxl
!pip install openpyxl

# Importamos las bibliotecas necesarias para trabajar con Excel
import openpyxl
from openpyxl.styles import Border, Side

# Solicitamos al usuario subir el archivo
uploaded = files.upload()
for fn in uploaded.keys():
    archivo = fn  # Guardamos el nombre del archivo subido

# Cargamos el archivo de Excel
wb = openpyxl.load_workbook(archivo)
hoja = wb.active

# Función para verificar si una celda tiene un borde a su derecha
def tiene_borde_derecho(celda):
    return celda.border.right.style is not None and celda.border.right.style != "none"

# Función para determinar si hay contenido en un rango de celdas
def hay_contenido_en_rango(inicio, fin):
    for col in range(inicio[1], fin[1]):
        if hoja.cell(row=inicio[0], column=col).value:
            return True
    return False

bordes_encontrados = []

# Identificamos las celdas con borde a la derecha y verificamos el contenido
for fila in hoja.iter_rows():
    for celda in fila:
        if tiene_borde_derecho(celda):
            bordes_encontrados.append(celda.coordinate)

            # Si tenemos dos bordes detectados
            if len(bordes_encontrados) == 2:
                inicio = openpyxl.utils.coordinate_to_tuple(bordes_encontrados[0])
                fin = openpyxl.utils.coordinate_to_tuple(bordes_encontrados[1])

                if fin[1] - inicio[1] == 1:  # El borde cubre solo una celda
                    if hoja.cell(row=inicio[0], column=inicio[1]).value:
                        print(f"En la celda {bordes_encontrados[0]} hay algo escrito.")
                elif hay_contenido_en_rango(inicio, fin):  # El borde cubre múltiples celdas
                    print(f"De la celda {bordes_encontrados[0]} a la celda {chr(ord(bordes_encontrados[1][0])-1) + str(inicio[0])} hay algo escrito.")

                # Limpiamos la lista de bordes para continuar con los siguientes
                bordes_encontrados.clear()

# No olvides cerrar el archivo de Excel después de leerlo
wb.close()

```
