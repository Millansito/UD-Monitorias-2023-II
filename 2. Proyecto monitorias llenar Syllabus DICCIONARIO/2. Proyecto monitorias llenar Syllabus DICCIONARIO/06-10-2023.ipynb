```python
from google.colab import files
import openpyxl
from openpyxl.styles import Border, Side
from openpyxl.utils import get_column_letter, column_index_from_string
import re

def tiene_borde_derecho(celda):
    return celda.border.right.style is not None and celda.border.right.style != "none"

def intervalo_sin_contenido(inicio, fin, sheet):
    fila = inicio[0]
    for col in range(inicio[1] + 1, fin[1]):
        if sheet.cell(row=fila, column=col).value:
            return False
    return True

def tiene_borde_con_contigua(celda, sheet):
    if celda.border.right.style:
        siguiente_col_index = celda.column + 1
        siguiente_celda = sheet.cell(row=celda.row, column=siguiente_col_index)
        if siguiente_celda.border.left.style:
            return True
    return False

def ordenar_por_coordenada(elemento):
    coord = re.search("[A-Z]+\d+", elemento).group()
    letra = re.search("[A-Z]+", coord).group()
    numero = int(re.search("\d+", coord).group())
    return (numero, column_index_from_string(letra))

uploaded = files.upload()
archivo = next(iter(uploaded))

wb = openpyxl.load_workbook(archivo)
hoja = wb.active

bordes_encontrados = []
intervalos_detectados = []
celdas_vacias_con_borde = []
salidas = []

for fila in hoja.iter_rows():
    for celda in fila:
        if tiene_borde_derecho(celda):
            bordes_encontrados.append(celda.coordinate)

            if len(bordes_encontrados) == 2:
                inicio = openpyxl.utils.coordinate_to_tuple(bordes_encontrados[0])
                fin = openpyxl.utils.coordinate_to_tuple(bordes_encontrados[1])

                if intervalo_sin_contenido(inicio, fin, hoja):
                    intervalos_detectados.append((bordes_encontrados[0], bordes_encontrados[1]))
                bordes_encontrados.clear()

        if not isinstance(celda, openpyxl.cell.cell.MergedCell) and (celda.value is None or celda.value == ""):
            if tiene_borde_con_contigua(celda, hoja):
                col_letter = get_column_letter(celda.column)
                celdas_vacias_con_borde.append(f'{col_letter}{celda.row}')

for inicio, fin in intervalos_detectados:
    if fin in celdas_vacias_con_borde:
        salidas.append(f'La celda {inicio} está vacía y tiene un borde con la celda contigua.')
    else:
        salidas.append(f"De la celda {inicio} a la celda {fin} no hay nada escrito.")

for celda in celdas_vacias_con_borde:
    salidas.append(f'La celda {celda} está vacía y tiene un borde con la celda contigua.')

salidas_ordenadas = sorted(salidas, key=ordenar_por_coordenada)

for salida in salidas_ordenadas:
    print(salida)

wb.close()

```
