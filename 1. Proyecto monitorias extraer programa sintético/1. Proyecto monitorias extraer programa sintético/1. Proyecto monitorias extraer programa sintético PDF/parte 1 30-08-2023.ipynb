```python
from PyPDF2 import PdfReader

def pdf_to_text(filename):
    with open(filename, 'rb') as file:
        reader = PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()  # Usa extract_text en lugar de extractText
    return text

pdf_filename = list(uploaded.keys())[0]
pdf_text = pdf_to_text(pdf_filename)
```

```python
import json
import re

# Carga el contenido del archivo JSON
with open('output.json', 'r') as json_file:
    data = json.load(json_file)

# Convierte los datos JSON en una cadena de texto
json_text = json.dumps(data)

# Divide la cadena en partes usando una expresi√≥n regular que busca palabras en may√∫sculas
split_parts = re.split(r'(?=[A-Z])', json_text)

# Filtra las partes vac√≠as y recorta los p√°rrafos
max_paragraph_length = 800  # Define la longitud m√°xima de los p√°rrafos
paragraphs = []
current_paragraph = ""
for part in split_parts:
    if part.strip():
        if len(current_paragraph) + len(part) <= max_paragraph_length:
            current_paragraph += part
        else:
            paragraphs.append(current_paragraph)
            current_paragraph = part
    elif current_paragraph:
        paragraphs.append(current_paragraph)
        current_paragraph = ""

# Imprime los p√°rrafos recortados
for index, paragraph in enumerate(paragraphs):
    print(f"P√°rrafo {index + 1}:")
    print(paragraph)
    print("-" * 50)
```

```python
import json
import os

# Cargar el archivo JSON
with open('output.json', 'r') as f:
    data = json.load(f)

# Crear una carpeta para los archivos de salida
output_folder = 'archivos_separados'
os.makedirs(output_folder, exist_ok=True)

# Iterar a trav√©s de los elementos del JSON
for clave, valor in data.items():
    # Buscar palabras en may√∫sculas en el valor
    palabras_mayusculas = [palabra for palabra in valor.split() if palabra.isupper()]

    # Crear un diccionario con los datos de la clave actual
    datos_clave = {
        clave: valor,
        "palabras_mayusculas": palabras_mayusculas
    }

    # Convertir el nombre de la clave a un formato adecuado para el nombre de archivo
    nombre_archivo = f"{clave.lower()}.json"

    # Guardar el diccionario como un archivo JSON en la carpeta de salida
    with open(os.path.join(output_folder, nombre_archivo), 'w') as f:
        json.dump(datos_clave, f, indent=4)

print("Archivos creados correctamente.")
print(datos_clave)
```
