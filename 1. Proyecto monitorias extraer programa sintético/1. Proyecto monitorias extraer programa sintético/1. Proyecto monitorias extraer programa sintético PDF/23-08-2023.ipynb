```python
import PyPDF2

# Nombre del archivo PDF y del archivo de salida TXT
input_pdf = "Poemas.pdf"
output_txt = "Poemas.txt"

# Abrir el archivo PDF en modo binario y leerlo
with open(input_pdf, "rb") as file:
    reader = PyPDF2.PdfReader(file)

    # Asegurarte de que el PDF no esté encriptado antes de leerlo
    if reader.is_encrypted:
        reader.decrypt("")

    # Extraer el texto de cada página y escribirlo en el archivo TXT
    with open(output_txt, "w", encoding="utf-8") as txt_out:
        for page in reader.pages:
            text = page.extract_text()
            txt_out.write(text + "\n")

print(f"Archivo {input_pdf} ha sido convertido a {output_txt}")
```

```python
import spacy

# Carga el modelo en español
nlp = spacy.load("es_core_news_sm")

# Leer el archivo "Poemas.txt"
with open("Poemas.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Procesar el texto con SpaCy
doc = nlp(text)


# Dividir el documento por párrafos. En este caso, consideraremos que un párrafo termina cuando hay un doble salto de línea.
paragraphs = text.split("\n\n")

# Procesar cada párrafo con SpaCy (opcional, si quieres hacer un análisis más profundo por párrafo)
for para in paragraphs:
    paragraph_doc = nlp(para)
    # Aquí puedes hacer análisis adicionales con el objeto paragraph_doc, como tokenización, detección de entidades, etc.
    print(paragraph_doc)
```
