```python
from PyPDF2 import PdfReader

def pdf_to_text(filename):
    with open(filename, 'rb') as file:
        reader = PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()  # Usa extract_text en lugar de extractText
    return text

pdf_filename = list(uploaded.keys())[0]
pdf_text = pdf_to_text(pdf_filename)
```

```python
import json

data = {
    "contenido": pdf_text
}

# Si deseas guardar el JSON en un archivo:
with open("output.json", "w") as json_file:
    json.dump(data, json_file)

# Mostrar el contenido en formato JSON:
print(json.dumps(data, indent=4))
```

```python
import json
import re

# Carga el contenido del archivo JSON
with open('output.json', 'r') as json_file:
    data = json.load(json_file)

# Convierte los datos JSON en una cadena de texto
json_text = json.dumps(data)

# Divide la cadena en partes usando una expresión regular que busca palabras en mayúsculas
split_parts = re.split(r'(?=[A-Z])', json_text)

# Filtra las partes vacías y recorta los párrafos
max_paragraph_length = 800  # Define la longitud máxima de los párrafos
paragraphs = []
current_paragraph = ""
for part in split_parts:
    if part.strip():
        if len(current_paragraph) + len(part) <= max_paragraph_length:
            current_paragraph += part
        else:
            paragraphs.append(current_paragraph)
            current_paragraph = part
    elif current_paragraph:
        paragraphs.append(current_paragraph)
        current_paragraph = ""

# Imprime los párrafos recortados
for index, paragraph in enumerate(paragraphs):
    print(f"Párrafo {index + 1}:")
    print(paragraph)
    print("-" * 50)
```
